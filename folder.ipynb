```python
import os
from pathlib import Path

# Define the folder and file structure
# Keys are directories, values are lists of files within them
structure = {
    "data/raw": [],
    "data/vector_store": [],
    "configs/prompts": ["rag_prompt.md", "hybrid_prompt.md"],
    "configs": ["config.yaml"],
    "src": [
        "__init__.py", "ingest.py", "chunking.py", "embedding.py", 
        "index.py", "retriever.py", "rag_pipeline.py", "router.py", "llm.py"
    ],
    "eval": ["questions.json", "run_eval.py"],
    "notebooks": [],
}

# Root level files
root_files = ["app.py", "requirements.txt", "README.md", ".env"]

def create_project_structure():
    # 1. Create directories and nested files
    for folder, files in structure.items():
        Path(folder).mkdir(parents=True, exist_ok=True)
        for file in files:
            file_path = Path(folder) / file
            if not file_path.exists():
                file_path.touch()
                print(f"Created file: {file_path}")

    # 2. Create root level files
    for file in root_files:
        file_path = Path(file)
        if not file_path.exists():
            file_path.touch()
            print(f"Created file: {file_path}")
            
    # 3. Optional: Add basic boilerplate to config.yaml
    config_path = Path("configs/config.yaml")
    if config_path.stat().st_size == 0:
        config_path.write_text("model:\n  name: 'gpt-4o'\n  temperature: 0\npaths:\n  raw_data: 'data/raw/'")

    print("\n✅ Project structure created successfully!")

create_project_structure()
```
